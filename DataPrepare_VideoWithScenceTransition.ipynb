```python
import json
import cv2
import numpy as np
import pandas as pd
import ffmpeg
from scenedetect import detect, ContentDetector
from tqdm import tqdm
import os

# 配置参数
CLIP_LENGTH = 10  # 目标片段时长（单位：秒）
PADDING = 5  # 片段前后扩展的秒数
MIN_SCENE_LENGTH = 3  # 场景片段最小长度
MAX_SCENE_LENGTH = 10  # 场景片段最大长度

def detect_scenes(video_path):
    """使用 PySceneDetect 检测视频中的转场时间点"""
    scene_list = detect(video_path, ContentDetector())
    return [(scene[1].get_seconds()) for scene in scene_list]  # 仅返回转场时间（秒）

def extract_transitional_clips(video_path, scene_timestamps):
    """基于转场时间点提取至少包含一帧转场的 10 秒视频片段"""
    video_name = os.path.basename(video_path).split('.')[0]
    output_clips = []

    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_duration = total_frames / fps  # 视频总时长（秒）

    for timestamp in scene_timestamps:
        start_time = max(0, timestamp - PADDING)
        end_time = min(video_duration, timestamp + PADDING)

        if end_time - start_time > MAX_SCENE_LENGTH:
            end_time = start_time + MAX_SCENE_LENGTH  # 确保不超过最大长度

        output_filename = f"{video_name}_{int(start_time)}-{int(end_time)}.mp4"
        output_path = os.path.join(OUTPUT_VIDEO_DIR, output_filename)

        # 执行视频裁剪
        (
            ffmpeg
            .input(video_path, ss=start_time, to=end_time)
            .output(output_path, vcodec="libx264", acodec="aac")
            .run(overwrite_output=True, quiet=True)
        )

        output_clips.append({
            "file_path": output_path,
            "video_name": video_name,
            "start_time": start_time,
            "end_time": end_time,
            "duration": end_time - start_time,
            "transition_frame": timestamp
        })

    cap.release()
    return output_clips

def validate_clips(clips):
    """确保所有片段包含至少一帧转场"""
    filtered_clips = []

    for clip in tqdm(clips, desc="Validating Clips"):
        cap = cv2.VideoCapture(clip["file_path"])
        prev_frame = None
        transition_detected = False

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if prev_frame is not None:
                diff = np.mean(cv2.absdiff(prev_frame, frame))  # 计算帧差
                if diff > 50:  # 经验阈值，较高的帧差意味着可能是转场
                    transition_detected = True
                    break

            prev_frame = frame

        cap.release()

        if transition_detected:
            filtered_clips.append(clip)  # 仅保留符合至少一帧转场的片段

    return filtered_clips

def save_metadata_to_json(filtered_clips):
    """导出符合条件的片段数据为 JSON 格式"""
    output_data = [
        {"file_path": clip["file_path"], "text": ""}
        for clip in filtered_clips
    ]

    with open(OUTPUT_JSON_FILE, "w", encoding="utf-8") as f:
        json.dump(output_data, f, ensure_ascii=False, indent=4)

    print(f"Metadata saved to {OUTPUT_JSON_FILE}")

def main():
    all_clips = []

    video_path = "/content/drive/MyDrive/Colab_Video_Processing/input_videos/-AM0oWK3RBA.mp4"
    scene_timestamps = detect_scenes(video_path)
    print("检测到的转场时间:", scene_timestamps)
    video_clips = extract_transitional_clips(video_path, scene_timestamps)
    all_clips.extend(video_clips)

    # STAGE 2: 确保片段符合“至少一帧转场”要求
    validated_clips = validate_clips(all_clips)

    # STAGE 3: 导出最终 JSON
    save_metadata_to_json(validated_clips)
if __name__ == "__main__":
    main()

```

```python
import json
import cv2
import numpy as np
import pandas as pd
import ffmpeg
from scenedetect import detect, ContentDetector
from tqdm import tqdm
import os
import signal

# 配置参数
CLIP_LENGTH = 10           # 目标片段时长（单位：秒）
PADDING = 5                # 片段前后扩展的秒数
MIN_SCENE_LENGTH = 3       # 场景片段最小长度
MAX_SCENE_LENGTH = 10      # 场景片段最大长度
MAX_PROCESSING_TIME = 300  # 单个视频处理的最大时间（单位：秒）
MAX_FILE_SIZE = 100 * 1024 * 1024  # 100 MB

# 全局变量，请根据实际路径进行设置
INPUT_VIDEO_DIR = "/content/drive/MyDrive/Colab_Video_Processing/input_videos"   # 输入视频文件夹路径
OUTPUT_VIDEO_DIR = "/content/drive/MyDrive/Colab_Video_Processing/output_clips"    # 输出 clip 的根文件夹路径
OUTPUT_JSON_FILE = "/content/drive/MyDrive/Colab_Video_Processing/output_meta.json"  # 输出 JSON 文件路径

# 检查 GPU 是否可用（需要支持 CUDA 的 OpenCV）
GPU_ENABLED = cv2.cuda.getCudaEnabledDeviceCount() > 0
if GPU_ENABLED:
    print("检测到 GPU，启用 GPU 加速。")
else:
    print("未检测到 GPU，使用 CPU 处理。")

# 定义超时异常及其处理函数
class TimeoutException(Exception):
    pass

def timeout_handler(signum, frame):
    raise TimeoutException("Processing time exceeded the limit.")

# 注册超时信号处理器（仅 Linux/Unix 环境支持）
signal.signal(signal.SIGALRM, timeout_handler)

def detect_scenes(video_path):
    """使用 PySceneDetect 检测视频中的转场时间点"""
    scene_list = detect(video_path, ContentDetector())
    return [scene[1].get_seconds() for scene in scene_list]  # 仅返回转场时间（秒）

def extract_transitional_clips(video_path, scene_timestamps):
    """基于转场时间点提取至少包含一帧转场的 10 秒视频片段，
       并将每个 clip 存放到以视频文件名命名的单独文件夹中"""
    video_name = os.path.basename(video_path).split('.')[0]
    output_clips = []

    # 获取视频信息
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_duration = total_frames / fps  # 视频总时长（秒）

    # 为当前视频创建单独的输出文件夹（以视频名命名）
    video_folder = os.path.join(OUTPUT_VIDEO_DIR, video_name)
    os.makedirs(video_folder, exist_ok=True)

    # 根据是否启用 GPU，选择不同的视频编码器
    if GPU_ENABLED:
        vcodec_param = "h264_nvenc"  # 使用 NVIDIA GPU 加速编码
    else:
        vcodec_param = "libx264"

    for timestamp in scene_timestamps:
        start_time = max(0, timestamp - PADDING)
        end_time = min(video_duration, timestamp + PADDING)

        if end_time - start_time > MAX_SCENE_LENGTH:
            end_time = start_time + MAX_SCENE_LENGTH  # 确保不超过最大长度

        output_filename = f"{video_name}_{int(start_time)}-{int(end_time)}.mp4"
        output_path = os.path.join(video_folder, output_filename)

        # 执行视频裁剪
        try:
            (
                ffmpeg
                .input(video_path, ss=start_time, to=end_time)
                .output(output_path, vcodec=vcodec_param, acodec="aac")
                .run(overwrite_output=True, quiet=True)
            )
        except Exception as e:
            print(f"裁剪视频 {video_path} 时出错: {e}")
            continue

        output_clips.append({
            "file_path": output_path,
            "video_name": video_name,
            "start_time": start_time,
            "end_time": end_time,
            "duration": end_time - start_time,
            "transition_frame": timestamp
        })

    cap.release()
    return output_clips

def validate_clips(clips):
    """确保所有片段包含至少一帧转场"""
    filtered_clips = []

    for clip in tqdm(clips, desc="Validating Clips"):
        cap = cv2.VideoCapture(clip["file_path"])
        transition_detected = False

        if GPU_ENABLED:
            gpu_prev = None
        else:
            prev_frame = None

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if GPU_ENABLED:
                # 上传当前帧到 GPU
                gpu_frame = cv2.cuda_GpuMat()
                gpu_frame.upload(frame)
                if gpu_prev is not None:
                    # 计算 GPU 上的帧差
                    gpu_diff = cv2.cuda.absdiff(gpu_prev, gpu_frame)
                    # 下载结果到 CPU 并计算均值
                    diff = np.mean(gpu_diff.download())
                else:
                    diff = 0
                gpu_prev = gpu_frame
            else:
                if prev_frame is not None:
                    diff = np.mean(cv2.absdiff(prev_frame, frame))
                else:
                    diff = 0
                prev_frame = frame

            # 根据经验阈值判断是否存在转场
            if diff > 50:
                transition_detected = True
                break

        cap.release()

        if transition_detected:
            filtered_clips.append(clip)  # 仅保留符合至少一帧转场的片段

    return filtered_clips

def save_metadata_to_json(filtered_clips):
    """导出符合条件的片段数据为 JSON 格式"""
    output_data = [
        {"file_path": clip["file_path"], "text": ""}
        for clip in filtered_clips
    ]

    with open(OUTPUT_JSON_FILE, "w", encoding="utf-8") as f:
        json.dump(output_data, f, ensure_ascii=False, indent=4)

    print(f"Metadata saved to {OUTPUT_JSON_FILE}")

def main():
    all_clips = []
    video_files = os.listdir(INPUT_VIDEO_DIR)
    for video_file in tqdm(video_files, desc="Processing Videos"):
        if not video_file.endswith((".mp4", ".avi", ".mov", ".mkv")):
            continue

        video_path = os.path.join(INPUT_VIDEO_DIR, video_file)

        # 检查文件大小，如果大于 100MB 则跳过
        file_size = os.path.getsize(video_path)
        if file_size > MAX_FILE_SIZE:
            print(f"跳过视频 {video_file}，文件大小 {file_size/1024/1024:.2f} MB 超过 100MB 限制。")
            continue

        print(f"Processing: {video_file}")

        try:
            # 为当前视频设置处理超时时间
            signal.alarm(MAX_PROCESSING_TIME)
            # STAGE 1: 解析转场点 & 生成片段
            scene_timestamps = detect_scenes(video_path)
            print("检测到的转场时间:", scene_timestamps)
            video_clips = extract_transitional_clips(video_path, scene_timestamps)
            all_clips.extend(video_clips)
        except TimeoutException:
            print(f"处理视频 {video_file} 超时，跳过。")
        except Exception as e:
            print(f"处理视频 {video_file} 时出现错误: {e}")
        finally:
            signal.alarm(0)  # 取消 alarm

    # STAGE 2: 确保片段符合“至少一帧转场”要求
    validated_clips = validate_clips(all_clips)

    # STAGE 3: 导出最终 JSON
    save_metadata_to_json(validated_clips)

if __name__ == "__main__":
    main()
```

```python
# %% [code]
# 导入必要的包
import os
import sys
import json
import cv2
import torch
import numpy as np
from PIL import Image
from lmdeploy import pipeline, TurbomindEngineConfig
from transformers import AutoTokenizer, AutoModelForCausalLM
import importlib
from tqdm import tqdm

# %% [code]
# 定义各模型的加载及处理函数

# 1. 加载 InternVL2 模型（视频描述生成）
def load_internvl2():
    # 使用 LMDeploy pipeline 加载 InternVL2 模型
    model_name = "internvl2_model"  # 模型下载至此目录
    backend_config = TurbomindEngineConfig(model_format='awq')  # 指定为AWQ量化模型
    pipe = pipeline(model_name, backend_config=backend_config, device_map="auto")
    return pipe

internvl2_pipe = load_internvl2()

# 2. 加载 Meta-Llama-3-8B-Instruct 模型（用于描述改写）
def load_llama():
    tokenizer = AutoTokenizer.from_pretrained("llama8b_model", use_fast=False)
    model = AutoModelForCausalLM.from_pretrained("llama8b_model", device_map="auto", torch_dtype=torch.float16)
    return tokenizer, model

llama_tokenizer, llama_model = load_llama()

# 3. 加载 VideoCLIP-XL 模型（用于视频与文本匹配过滤）
# 假设 VideoCLIP-XL 的代码和模型权重已下载到 "videoclip_xl_model" 目录中
# 这里需要将模型代码所在目录添加到 sys.path 中
videoclip_model_dir = os.path.abspath("videoclip_xl_model")
if videoclip_model_dir not in sys.path:
    sys.path.append(videoclip_model_dir)
# 假设 VideoCLIP-XL 的代码中定义了 VideoCLIP_XL 类，以及文本编码函数 tokenize() 位于 utils/text_encoder/text_encoder.py 中
try:
    from modeling import VideoCLIP_XL
except ImportError:
    raise ImportError("请检查 videoclip_xl_model 目录下是否存在 VideoCLIP-XL 的代码文件。")

def load_videoclip_xl():
    model = VideoCLIP_XL()
    # 加载权重，假设权重文件为 VideoCLIP-XL.bin
    weight_path = os.path.join(videoclip_model_dir, "VideoCLIP-XL.bin")
    state_dict = torch.load(weight_path, map_location="cpu")
    model.load_state_dict(state_dict)
    model = model.cuda().eval()
    return model

videoclip_xl = load_videoclip_xl()

# %% [code]
# 定义辅助函数：采样视频帧
def sample_frames_from_video(video_path, num_frames=8, target_size=(448, 448)):
    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames <= 0:
        cap.release()
        return frames
    step = max(1, total_frames // num_frames)
    for i in range(num_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * step)
        ret, frame = cap.read()
        if not ret:
            break
        # 转换 BGR -> RGB 并调整尺寸
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame_resized = cv2.resize(frame_rgb, target_size)
        frames.append(Image.fromarray(frame_resized))
    cap.release()
    return frames

# %% [code]
# 使用 InternVL2 生成视频描述
def generate_initial_caption(video_path, prompt="Describe the video in detail."):
    frames = sample_frames_from_video(video_path, num_frames=8, target_size=(448, 448))
    if len(frames) == 0:
        return ""
    # 使用 InternVL2 pipeline，输入为 (prompt, [帧列表])
    outputs = internvl2_pipe([(prompt, frames)])
    caption = outputs[0].text
    return caption

# %% [code]
# 使用 Llama 对描述进行改写
def rewrite_caption(initial_caption):
    instruction = (
        "You are an expert video script writer. Rewrite the following video description "
        "to be more detailed, vivid and suitable as a prompt for video generation.\n"
        f"Original description: {initial_caption}\n"
        "Rewritten description:"
    )
    inputs = llama_tokenizer(instruction, return_tensors="pt").to(llama_model.device)
    outputs = llama_model.generate(**inputs, max_new_tokens=100, temperature=0.7, do_sample=False)
    rewritten_text = llama_tokenizer.decode(outputs[0], skip_special_tokens=True)
    # 提取“Rewritten description:”后面的部分（若有）
    if "Rewritten description:" in rewritten_text:
        rewritten_text = rewritten_text.split("Rewritten description:")[-1].strip()
    return rewritten_text

# %% [code]
# 定义 VideoCLIP-XL 的预处理及相似度计算函数
def get_video_text_similarity(video_path, text):
    # 采样视频帧用于 VideoCLIP-XL（调整为224x224）
    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    num_frames = 8
    step = max(1, total_frames // num_frames)
    count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if count % step == 0:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_resized = cv2.resize(frame_rgb, (224, 224))
            frames.append(frame_resized)
        count += 1
        if len(frames) >= num_frames:
            break
    cap.release()
    if len(frames) == 0:
        return 0.0
    # 将帧归一化后堆叠为张量（1, T, 3, H, W）
    v_mean = np.array([0.485, 0.456, 0.406]).reshape(1,1,3)
    v_std = np.array([0.229, 0.224, 0.225]).reshape(1,1,3)
    frame_tensors = []
    for fr in frames:
        fr_norm = (fr / 255.0 - v_mean) / v_std
        fr_norm = fr_norm.astype('float32')
        # 扩展维度并调整形状为 (1, 3, H, W)
        fr_tensor = torch.from_numpy(np.transpose(fr_norm, (2, 0, 1)))
        frame_tensors.append(fr_tensor.unsqueeze(0))
    vid_tensor = torch.cat(frame_tensors, dim=0).unsqueeze(0).cuda()  # (1, T, 3, H, W)

    # 通过 VideoCLIP-XL 提取视频特征
    with torch.no_grad():
        video_features = videoclip_xl.vision_model.encode_video(vid_tensor).float()
        video_features = video_features / video_features.norm(dim=-1, keepdim=True)
        # 文本编码：假设 VideoCLIP-XL 提供了文本编码接口 tokenize() 位于 utils/text_encoder/text_encoder.py
        try:
            text_tokenizer = importlib.import_module("utils.text_encoder.text_encoder")
            text_inputs = text_tokenizer.tokenize([text], truncate=True).cuda()
        except Exception as e:
            print("加载 VideoCLIP-XL 文本编码器出错：", e)
            return 0.0
        text_features = videoclip_xl.text_model.encode_text(text_inputs).float()
        text_features = text_features / text_features.norm(dim=-1, keepdim=True)
        # 计算余弦相似度
        sim_score = float((text_features @ video_features.T).item())
    return sim_score

# %% [code]
# 主流程：读取所有视频片段，生成描述、改写并过滤，最后输出 JSON 文件
# 输入文件夹（含子文件夹）
INPUT_CLIPS_DIR = "/content/drive/MyDrive/Colab_Video_Processing/output_clips"
# 输出 JSON 文件路径（建议与 README 中保持一致）
OUTPUT_JSON_PATH = "/content/drive/MyDrive/Colab_Video_Processing/video_caption/datasets/panda_70m/videos_clips/meta_train_info.json"
os.makedirs(os.path.dirname(OUTPUT_JSON_PATH), exist_ok=True)

# 设置视频与文本匹配的最低阈值（可根据实际情况调整）
SIMILARITY_THRESHOLD = 0.20

final_video_captions = []

# 遍历所有子目录和文件（支持 .mp4, .avi, .mov, .mkv 等）
for root, dirs, files in os.walk(INPUT_CLIPS_DIR):
    for file in files:
        if file.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
            clip_path = os.path.join(root, file)
            print(f"Processing clip: {clip_path}")
            try:
                # 1. 使用 InternVL2 生成初始描述
                initial_caption = generate_initial_caption(clip_path)
                if not initial_caption:
                    print("未生成初始描述，跳过。")
                    continue
                print("Initial caption:", initial_caption)

                # 2. 使用 Llama 模型对描述进行改写
                rewritten_caption = rewrite_caption(initial_caption)
                print("Rewritten caption:", rewritten_caption)

                # 3. 利用 VideoCLIP-XL 模型计算视频与描述的相似度
                sim_score = get_video_text_similarity(clip_path, rewritten_caption)
                print(f"Video-Text Similarity Score: {sim_score:.4f}")

                # 4. 如果匹配分数达到要求，则保存结果
                if sim_score >= SIMILARITY_THRESHOLD:
                    final_video_captions.append({
                        "file_path": clip_path,
                        "caption": rewritten_caption
                    })
                else:
                    print("相似度较低，过滤掉此样本。")
            except Exception as e:
                print(f"处理 {clip_path} 时出错: {e}")

# 保存最终 JSON 文件
with open(OUTPUT_JSON_PATH, "w", encoding="utf-8") as f:
    json.dump(final_video_captions, f, ensure_ascii=False, indent=4)
print(f"输出文件已保存: {OUTPUT_JSON_PATH}")
```

```python
# %% [code]
# 导入必要的包
import os
import sys
import json
import cv2
import torch
import numpy as np
from PIL import Image
from transformers import pipeline
from tqdm import tqdm
import signal

# PySceneDetect 相关导入
from scenedetect import VideoManager, SceneManager
from scenedetect.detectors import ContentDetector

# 定义超时异常及处理器（防止单个视频处理时间过长）
class TimeoutException(Exception):
    pass

def timeout_handler(signum, frame):
    raise TimeoutException("Processing time exceeded the limit.")

# 注册超时信号处理器
signal.signal(signal.SIGALRM, timeout_handler)

# %% [code]
# 1. 加载图像字幕模型（用于生成视频描述）
def load_caption_model():
    # 使用 Hugging Face 的 pipeline 加载 BLIP 图像字幕模型
    # device=0 表示使用第一块 GPU（若无 GPU，可改为 -1 使用 CPU）
    captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base", device=0)
    return captioner

captioner = load_caption_model()

# %% [code]
# 使用 PySceneDetect 检测视频中的转场时间点
def detect_scenes(video_path):
    """
    使用 PySceneDetect 检测视频中的转场时间点，
    返回一个列表，每个元素为转场结束时刻（秒）。
    """
    video_manager = VideoManager([video_path])
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector())
    video_manager.set_downscale_factor()  # 默认下采样因子
    video_manager.start()
    scene_manager.detect_scenes(frame_source=video_manager)
    scene_list = scene_manager.get_scene_list()
    video_manager.release()
    # 返回每个场景的结束时刻（秒）
    return [scene[1].get_seconds() for scene in scene_list]

# %% [code]
# 定义辅助函数：均匀采样视频帧
def sample_frames_from_video(video_path, num_frames=8, target_size=(384,384)):
    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames <= 0:
        cap.release()
        return frames
    step = max(1, total_frames // num_frames)
    for i in range(num_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * step)
        ret, frame = cap.read()
        if not ret:
            break
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame_resized = cv2.resize(frame_rgb, target_size)
        frames.append(Image.fromarray(frame_resized))
    cap.release()
    return frames

# %% [code]
# 定义辅助函数：从转场前后采样帧
def sample_frames_around_transition(video_path, transition_time, n_before=4, n_after=4, target_size=(384,384)):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    transition_frame = int(transition_time * fps)
    # 采样转场前的帧（取紧靠转场前的 n_before 帧）
    before_indices = list(range(max(0, transition_frame - n_before), transition_frame))
    # 采样转场后的帧（从转场帧开始的 n_after 帧）
    after_indices = list(range(transition_frame, min(total_frames, transition_frame + n_after)))
    before_frames = []
    after_frames = []
    current_frame = 0
    ret, frame = cap.read()
    while ret:
        if current_frame in before_indices:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_resized = cv2.resize(frame_rgb, target_size)
            before_frames.append(Image.fromarray(frame_resized))
        if current_frame in after_indices:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame_resized = cv2.resize(frame_rgb, target_size)
            after_frames.append(Image.fromarray(frame_resized))
        current_frame += 1
        ret, frame = cap.read()
    cap.release()
    return before_frames, after_frames

# %% [code]
# 定义函数：生成单个视频的描述
def generate_caption(video_path):
    try:
        scenes = detect_scenes(video_path)
    except Exception as e:
        print(f"Scene detection error for {video_path}: {e}")
        scenes = []
    if scenes:
        # 如果检测到转场，取第一个转场时间
        transition_time = scenes[0]
        before_frames, after_frames = sample_frames_around_transition(video_path, transition_time)
        if len(before_frames) == 0 or len(after_frames) == 0:
            # 如果采样失败，则退回均匀采样
            frames = sample_frames_from_video(video_path, num_frames=8, target_size=(384,384))
            if not frames:
                return ""
            mid_frame = frames[len(frames)//2]
            result = captioner(mid_frame)
            caption = result[0]["generated_text"] if result and isinstance(result, list) and "generated_text" in result[0] else ""
            return caption
        # 分别从转场前后取中间帧生成描述
        mid_before = before_frames[len(before_frames)//2]
        mid_after = after_frames[len(after_frames)//2]
        result1 = captioner(mid_before)
        result2 = captioner(mid_after)
        caption1 = result1[0]["generated_text"] if result1 and isinstance(result1, list) and "generated_text" in result1[0] else ""
        caption2 = result2[0]["generated_text"] if result2 and isinstance(result2, list) and "generated_text" in result2[0] else ""
        # 按照格式拼接描述：前一个场景为 xxx，后一个场景为 xxx
        merged_caption = f"previous scene: {caption1.strip()}; then next scence: {caption2.strip()}"
        return merged_caption
    else:
        # 如果未检测到转场，则均匀采样8帧，取中间帧生成描述
        frames = sample_frames_from_video(video_path, num_frames=8, target_size=(384,384))
        if not frames:
            return ""
        mid_frame = frames[len(frames)//2]
        result = captioner(mid_frame)
        caption = result[0]["generated_text"] if result and isinstance(result, list) and "generated_text" in result[0] else ""
        return caption

# %% [code]
# 主流程：遍历视频文件，对每个视频生成描述并保存单独的 JSON 文件及文件夹，最后生成总的 meta JSON
INPUT_CLIPS_DIR = "/content/drive/MyDrive/Colab_Video_Processing/output_clips"  # 视频所在目录
OUTPUT_BASE_DIR = "/content/drive/MyDrive/Colab_Video_Processing/video_caption/segments"  # 各视频文件夹存放目录
META_JSON_PATH = "/content/drive/MyDrive/Colab_Video_Processing/video_caption/datasets/panda_70m/videos_clips/meta_train_info.json"

os.makedirs(OUTPUT_BASE_DIR, exist_ok=True)
os.makedirs(os.path.dirname(META_JSON_PATH), exist_ok=True)

meta_data = []

for root, dirs, files in os.walk(INPUT_CLIPS_DIR):
    for file in files:
        if file.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
            video_path = os.path.join(root, file)
            print(f"Processing video: {video_path}")
            try:
                # 设置超时（例如 120 秒），防止单个视频处理时间过长
                signal.alarm(120)
                caption = generate_caption(video_path)
                signal.alarm(0)  # 处理完成后取消超时提醒
                if not caption:
                    print("No caption generated, skipping.")
                    continue
                print("Caption:", caption)

                # 为每个视频创建输出文件夹，文件夹名称取视频文件名（去除扩展名）
                video_basename = os.path.splitext(os.path.basename(video_path))[0]
                video_output_dir = os.path.join(OUTPUT_BASE_DIR, video_basename)
                os.makedirs(video_output_dir, exist_ok=True)

                # 根据是否检测到转场，记录采样方式
                scene_flag = "transition" if detect_scenes(video_path) else "uniform"
                video_json = {
                    "video_path": video_path,
                    "caption": caption,
                    "segment_type": scene_flag
                }
                json_filename = os.path.join(video_output_dir, f"{video_basename}.json")
                with open(json_filename, "w", encoding="utf-8") as f:
                    json.dump(video_json, f, ensure_ascii=False, indent=4)

                meta_data.append({
                    "video_path": video_path,
                    "caption": caption,
                    "json_path": json_filename,
                    "output_folder": video_output_dir
                })
            except Exception as e:
                print(f"Error processing {video_path}: {e}")

# 保存总的 meta JSON 文件
with open(META_JSON_PATH, "w", encoding="utf-8") as f:
    json.dump(meta_data, f, ensure_ascii=False, indent=4)
print(f"Meta JSON saved: {META_JSON_PATH}")
```
